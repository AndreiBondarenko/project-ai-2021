```python
import utils
import numpy as np
from tqdm.auto import tqdm
import scipy.sparse as sps
import matplotlib.pyplot as plt
from recommenders.mf import MatrixFactorization
%matplotlib inline
```

```python
x = sps.load_npz('./data/train.npz')
y = sps.load_npz('./data/test.npz')
_y = y.tocsr()
```

```python
for K in [2**i for i in range(3, 10)]:
    print(f'\nK = {K}')
    recommender = MatrixFactorization(K=K, iterations=300, gamma=0.3)
    e_train, e_test = recommender.train(x, y)
    plt.plot(e_train, label="train")
    plt.plot(e_test, label="test")
    plt.xlabel("iteration")
    plt.ylabel("MSE")
    plt.legend()
    plt.show()
```
